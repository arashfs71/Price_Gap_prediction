```python
import pandas as pd
```

```python
def construct_DAM_dataframe(path):
    dam_data = pd.read_csv(path)
    dam_data = dam_data.drop(columns=['OPR_INTERVAL', 'GRP_TYPE', 'NODE_ID_XML', 'NODE_ID', 'PNODE_RESMRID'])
    dam_data =  dam_data.rename(columns={"INTERVALSTARTTIME_GMT": "start_datetime", "INTERVALENDTIME_GMT":"end_datetime", "OPR_HR": "hr_index", "OPR_DT":"date", "NODE":"node", "MARKET_RUN_ID":"market","LMP_TYPE":"lmp_type","MW":"dam_price_per_mwh"})
    df = dam_data.pivot_table(index=['date', 'hr_index', 'node'], columns='XML_DATA_ITEM', values='dam_price_per_mwh').reset_index().rename(columns={'LMP_CONG_PRC': 'DAM LMP_CONG_PRC', 'LMP_ENE_PRC': 'DAM LMP_ENE_PRC', 'LMP_LOSS_PRC':'DAM LMP_LOSS_PRC', 'LMP_PRC': 'DAM LMP_PRC'})
    df.columns.name=None
    return df
```

```python
def construct_HASP_dataframe(path):
    hasp_data = pd.read_csv(path)
    hasp_data = hasp_data.drop(columns=['GRP_TYPE', 'NODE_ID_XML', 'NODE_ID', 'PNODE_RESMRID'])
    hasp_data =  hasp_data.rename(columns={"INTERVALSTARTTIME_GMT": "start_datetime", "INTERVALENDTIME_GMT":"end_datetime", "OPR_HR": "hr_index", "OPR_DT":"date", "NODE":"node", "MARKET_RUN_ID":"market","LMP_TYPE":"lmp_type","MW":"hasp_price_per_mwh"})
    df = hasp_data.pivot_table(index=['date', 'hr_index', 'OPR_INTERVAL', 'node'], columns='XML_DATA_ITEM', values='hasp_price_per_mwh').reset_index().rename(columns={'LMP_CONG_PRC': 'HASP LMP_CONG_PRC', 'LMP_ENE_PRC': 'HASP LMP_ENE_PRC', 'LMP_LOSS_PRC':'HASP LMP_LOSS_PRC', 'LMP_PRC': 'HASP LMP_PRC', 'LMP_GHG_PRC':'HASP LMP_GHG_PRC'})
    df.columns.name=None
    df = df.pivot_table(index=['date', 'hr_index', 'node'], columns='OPR_INTERVAL', values=['HASP LMP_ENE_PRC', 'HASP LMP_ENE_PRC', 'HASP LMP_GHG_PRC', 'HASP LMP_LOSS_PRC', 'HASP LMP_PRC']).reset_index()
    df.columns.name=None
    return df
```
