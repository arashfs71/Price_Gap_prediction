```python
import pandas as pd
import os
from pandas.io.json import json_normalize
import json
```

```python
li = []
for filename in os.listdir('C:\\Users\\Nika\\PycharmProjects\\PricePrediction\\caiso_downloads\\DAM_unziped'):
    path = 'C:\\Users\\Nika\\PycharmProjects\\PricePrediction\\caiso_downloads\\DAM_unziped\\'+filename
    df = pd.read_csv(path, index_col=None, header=0)
    li.append(df)
dam_df = pd.concat(li, axis=0, ignore_index=True)
```

```python
dam_df = dam_df.drop(columns=['OPR_INTERVAL', 'GRP_TYPE', 'NODE_ID_XML', 'NODE_ID', 'PNODE_RESMRID'])
dam_df =  dam_df.rename(columns={"INTERVALSTARTTIME_GMT": "start_datetime", "INTERVALENDTIME_GMT":"end_datetime", "OPR_HR": "hr_index", "OPR_DT":"date", "NODE":"node", "MARKET_RUN_ID":"market","LMP_TYPE":"lmp_type","MW":"dam_price_per_mwh"})
dam_df = dam_df.pivot_table(index=['date', 'hr_index', 'node'], columns='XML_DATA_ITEM', values='dam_price_per_mwh').reset_index().rename(columns={'LMP_CONG_PRC': 'DAM LMP_CONG_PRC', 'LMP_ENE_PRC': 'DAM LMP_ENE_PRC', 'LMP_LOSS_PRC':'DAM LMP_LOSS_PRC', 'LMP_PRC': 'DAM LMP_PRC'})
dam_df.columns.name=None
```
