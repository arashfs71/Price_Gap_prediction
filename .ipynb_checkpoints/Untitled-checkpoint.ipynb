```python
import pandas as pd
import os
from pandas.io.json import json_normalize
import json
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
fr_df = pd.read_csv('C:\\Users\\Nika\\Desktop\\srilm-1.7.1-20200225T011940Z-001\\srilm-1.7.1\\bin\\macosx\\data\\French_dist.csv', header=None)
en_df = pd.read_csv('C:\\Users\\Nika\\Desktop\\srilm-1.7.1-20200225T011940Z-001\\srilm-1.7.1\\bin\\macosx\\data\\English_dist.csv', header=None)
it_df = pd.read_csv('C:\\Users\\Nika\\Desktop\\srilm-1.7.1-20200225T011940Z-001\\srilm-1.7.1\\bin\\macosx\\data\\Italian.csv', header=None)
ru_df = pd.read_csv('C:\\Users\\Nika\\Desktop\\srilm-1.7.1-20200225T011940Z-001\\srilm-1.7.1\\bin\\macosx\\data\\Russian.csv', header=None)
es_df = pd.read_csv('C:\\Users\\Nika\\Desktop\\srilm-1.7.1-20200225T011940Z-001\\srilm-1.7.1\\bin\\macosx\\data\\Spanish.csv', header=None)
zh_df = pd.read_csv('C:\\Users\\Nika\\Desktop\\srilm-1.7.1-20200225T011940Z-001\\srilm-1.7.1\\bin\\macosx\\data\\Chinese_dist.csv', header=None)
```

```python
fr_df = fr_df.transpose()
en_df = en_df.transpose()
it_df = it_df.transpose()
ru_df = ru_df.transpose()
es_df = es_df.transpose()
zh_df = zh_df.transpose()
```

```python
fr_df = fr_df.rename({0: 'Year', 1: 'Tokens'}, axis='columns')
en_df = en_df.rename({0: 'Year', 1: 'Tokens'}, axis='columns') 
it_df = it_df.rename({0: 'Year', 1: 'Tokens'}, axis='columns')
ru_df = ru_df.rename({0: 'Year', 1: 'Tokens'}, axis='columns')
es_df = es_df.rename({0: 'Year', 1: 'Tokens'}, axis='columns')
zh_df = zh_df.rename({0: 'Year', 1: 'Tokens'}, axis='columns')
```

```python
fr_df = fr_df[fr_df.Year <= 1930]
fr_df = fr_df[fr_df.Year >= 1820]

en_df = en_df[en_df.Year <= 1930]
en_df = en_df[en_df.Year >= 1820]

it_df = it_df[it_df.Year <= 1930]
it_df = it_df[it_df.Year >= 1820]

ru_df = ru_df[ru_df.Year <= 1930]
ru_df = ru_df[ru_df.Year >= 1820]

es_df = es_df[es_df.Year <= 1930]
es_df = es_df[es_df.Year >= 1820]

zh_df = zh_df[zh_df.Year <= 1930]
zh_df = zh_df[zh_df.Year >= 1820]

```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(18, 6), dpi=80, facecolor='w', edgecolor='k')

plt.plot(fr_df.Year, fr_df.Tokens, color='cyan', label="French")
plt.plot(en_df.Year, en_df.Tokens, color='green', label="English")
plt.plot(it_df.Year, it_df.Tokens, color='blue', label="Italian")
plt.plot(es_df.Year, es_df.Tokens, color='yellow', label="Spanish")
plt.plot(ru_df.Year, ru_df.Tokens, color='black', label="Russian")
plt.plot(zh_df.Year, zh_df.Tokens, color='red',label="Chinese")
plt.legend(loc="upper left")
plt.show()
```
